Ye notebook aapko Chinese video se audio extract karke uska text (script) nikaalne me madad karega using Whisper.

✅ Whisper model se Chinese script

```python
# 📁 Video Upload karo
from google.colab import files
uploaded = files.upload()
video_file = list(uploaded.keys())[0]
print("Uploaded:", video_file)
```

```python
# 🎧 Video se Audio Extract karo
from moviepy.editor import VideoFileClip
video = VideoFileClip(video_file)
audio_file = "audio.wav"
video.audio.write_audiofile(audio_file)
print("Audio extracted as:", audio_file)
```

```python
# 🤖 Whisper Model se Chinese Script nikaalo
import whisper
model = whisper.load_model("base")  # base ya small le sakte ho
result = model.transcribe(audio_file, language="zh")
print("\n📜 Transcript:\n")
print(result["text"])
```

```python
# 💾 Script File Save aur Download karo
with open("transcript.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

files.download("transcript.txt")
```
